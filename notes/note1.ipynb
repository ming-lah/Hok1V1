### **注意内容**

需要掌握算法前提：PPO算法，复赛使用的都是PPO算法，可以先提前熟悉一下，同时可以上网搜索相关的论文(1V1环境)

框架和初赛时基本类似，主要工作还是：特征处理，网络设计，奖励设计三个方面

比赛资源：本次是在线开发(没有本地副本)，所以可能需要大家每次修改完之后手动复制保存到本地(可能会有其他比较好的方法解决)，然后在群内讨论/上传github。训练资源由平台提供，每个团队同时只能训练一个模型，而且根据其他的经验，训练时间可能比较久，所以真正进入比赛开发后需要慎重一些。同时评估资源也有限制，一共50000局的评估(好像会刷新每周？)。评估时可以看到具体的回放，需要下载平台(赛题指引有说明)
